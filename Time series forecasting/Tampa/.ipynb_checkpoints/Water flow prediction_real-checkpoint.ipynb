```python
from statsmodels.tsa.stattools import adfuller

# Perform Augmented Dickey-Fuller test
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print(f'ADF Statistic: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical Values: {result[4]}')
    if result[1] <= 0.05:
        print('Strong evidence against the null hypothesis, reject the null hypothesis. Data has no unit root and is stationary')
    else:
        print('Weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ')

# Apply the test on your time series data
adf_test(data['TotalTreated water leaving plant,MG'])
adf_test(data['Adjusted Daily Consumption, Kwh'])
```

Random Forest as the machine learning model and SARIMA as the time series model. It then averages the predictions from both models to create an ensemble. You can adjust the ensemble weights based on the performance of individual models.

```python
# Importing required libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split
from statsmodels.tsa.statespace.sarimax import SARIMAX
from math import sqrt
```

```python
# Assuming 'data' is your DataFrame containing historical data
# Log transformation of the target variables
data['Log_TotalTreated'] = np.log1p(data['TotalTreated water leaving plant,MG'])
data['Log_EnergyConsumption'] = np.log1p(data['Adjusted Daily Consumption, Kwh'])

# Feature Engineering: Create lag features for target variables
for col in ['Log_EnergyConsumption']:
    for lag in range(1, 4):  # Create 3 lag features
        data[f'{col}_lag{lag}'] = data[col].shift(lag)

# Drop NaN values created due to lag features
data.dropna(inplace=True)
```

```python
# Train-Test Split
train_size = int(len(data) * 0.8)
train, test = data[:train_size], data[train_size:]

# Prepare data for machine learning models for actual values
X_train_actual = train.drop(['Date', 'TotalTreated water leaving plant,MG', 'Adjusted Daily Consumption, Kwh'], axis=1)
y_train_treated_actual = train['TotalTreated water leaving plant,MG']
y_train_energy_actual = train['Adjusted Daily Consumption, Kwh']

X_test_actual = test.drop(['Date', 'TotalTreated water leaving plant,MG', 'Adjusted Daily Consumption, Kwh'], axis=1)
y_test_treated_actual = test['TotalTreated water leaving plant,MG']
y_test_energy_actual = test['Adjusted Daily Consumption, Kwh']

# Prepare data for machine learning models for log-transformed values
X_train_log = train.drop(['Date', 'Log_TotalTreated', 'Log_EnergyConsumption'], axis=1)
y_train_treated_log = train['Log_TotalTreated']
y_train_energy_log = train['Log_EnergyConsumption']

X_test_log = test.drop(['Date', 'Log_TotalTreated', 'Log_EnergyConsumption'], axis=1)
y_test_treated_log = test['Log_TotalTreated']
y_test_energy_log = test['Log_EnergyConsumption']
```

```python
# Machine Learning Model Training for Treated Water
rf_model_treated = RandomForestRegressor()
rf_model_treated.fit(X_train_actual, y_train_treated_actual)

gb_model_treated = GradientBoostingRegressor()
gb_model_treated.fit(X_train_actual, y_train_treated_actual)

# Machine Learning Model Training for Log-Transformed Treated Water
rf_model_Logtreated = RandomForestRegressor()
rf_model_Logtreated.fit(X_train_log, y_train_treated_log)

gb_model_Logtreated = GradientBoostingRegressor()
gb_model_Logtreated.fit(X_train_log, y_train_treated_log)

# Machine Learning Model Training for Energy Consumption
rf_model_energy = RandomForestRegressor()
rf_model_energy.fit(X_train_actual, y_train_energy_actual)

gb_model_energy = GradientBoostingRegressor()
gb_model_energy.fit(X_train_actual, y_train_energy_actual)

# Machine Learning Model Training for Log-Transformed Energy Consumption
rf_model_log_energy = RandomForestRegressor()
rf_model_log_energy.fit(X_train_log, y_train_energy_log)

gb_model_log_energy = GradientBoostingRegressor()
gb_model_log_energy.fit(X_train_log, y_train_energy_log)

# SARIMA Model Training for Treated Water
sarima_model_treated = SARIMAX(y_train_treated_actual, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit_treated = sarima_model_treated.fit(disp=False)

# SARIMA Model Training for Log-Transformed Treated Water
sarima_model_Logtreated = SARIMAX(y_train_treated_log, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit_Logtreated = sarima_model_Logtreated.fit(disp=False)

# SARIMA Model Training for Energy Consumption
sarima_model_energy = SARIMAX(y_train_energy_actual, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit_energy = sarima_model_energy.fit(disp=False)

# SARIMA Model Training for Log-Transformed Energy Consumption
sarima_model_log_energy = SARIMAX(y_train_energy_log, order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_fit_log_energy = sarima_model_log_energy.fit(disp=False)
```

```python
# Making Predictions for Actual Values
rf_pred_treated_actual = rf_model_treated.predict(X_test_actual)
gb_pred_treated_actual = gb_model_treated.predict(X_test_actual)
rf_pred_energy_actual = rf_model_energy.predict(X_test_actual)
gb_pred_energy_actual = gb_model_energy.predict(X_test_actual)

sarima_pred_treated_actual = sarima_fit_treated.predict(len(y_train_treated_actual), len(data) - 1)
sarima_pred_energy_actual = sarima_fit_energy.predict(len(y_train_energy_actual), len(data) - 1)

# Ensemble Predictions for Actual Values
ensemble_pred_treated_actual = 0.25 * (rf_pred_treated_actual + gb_pred_treated_actual + sarima_pred_treated_actual)
ensemble_pred_energy_actual = 0.25 * (rf_pred_energy_actual + gb_pred_energy_actual + sarima_pred_energy_actual)

# Making Predictions for Log-Transformed Values
rf_pred_treated_log = rf_model_Logtreated.predict(X_test_log)
gb_pred_treated_log = gb_model_Logtreated.predict(X_test_log)
rf_pred_energy_log = rf_model_log_energy.predict(X_test_log)
gb_pred_energy_log = gb_model_log_energy.predict(X_test_log)

sarima_pred_treated_log = sarima_fit_Logtreated.predict(len(y_train_treated_log), len(data) - 1)
sarima_pred_energy_log = sarima_fit_log_energy.predict(len(y_train_energy_log), len(data) - 1)

# Ensemble Predictions for Log-Transformed Values
ensemble_pred_treated_log = 0.25 * (rf_pred_treated_log + gb_pred_treated_log + sarima_pred_treated_log)
ensemble_pred_energy_log = 0.25 * (rf_pred_energy_log + gb_pred_energy_log + sarima_pred_energy_log)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from math import sqrt

# Compute RMSE and MAE for Actual Values
# Treated Water
print("=== Actual Values for Treated Water ===")
print("RF RMSE:", sqrt(mean_squared_error(y_test_treated_actual, rf_pred_treated_actual)))
print("GB RMSE:", sqrt(mean_squared_error(y_test_treated_actual, gb_pred_treated_actual)))
print("SARIMA RMSE:", sqrt(mean_squared_error(y_test_treated_actual, sarima_pred_treated_actual)))
print("Ensemble RMSE:", sqrt(mean_squared_error(y_test_treated_actual, ensemble_pred_treated_actual)))

print("RF MAE:", mean_absolute_error(y_test_treated_actual, rf_pred_treated_actual))
print("GB MAE:", mean_absolute_error(y_test_treated_actual, gb_pred_treated_actual))
print("SARIMA MAE:", mean_absolute_error(y_test_treated_actual, sarima_pred_treated_actual))
print("Ensemble MAE:", mean_absolute_error(y_test_treated_actual, ensemble_pred_treated_actual))

# Energy Consumption
print("\n=== Actual Values for Energy Consumption ===")
print("RF RMSE:", sqrt(mean_squared_error(y_test_energy_actual, rf_pred_energy_actual)))
print("GB RMSE:", sqrt(mean_squared_error(y_test_energy_actual, gb_pred_energy_actual)))
print("SARIMA RMSE:", sqrt(mean_squared_error(y_test_energy_actual, sarima_pred_energy_actual)))
print("Ensemble RMSE:", sqrt(mean_squared_error(y_test_energy_actual, ensemble_pred_energy_actual)))

print("RF MAE:", mean_absolute_error(y_test_energy_actual, rf_pred_energy_actual))
print("GB MAE:", mean_absolute_error(y_test_energy_actual, gb_pred_energy_actual))
print("SARIMA MAE:", mean_absolute_error(y_test_energy_actual, sarima_pred_energy_actual))
print("Ensemble MAE:", mean_absolute_error(y_test_energy_actual, ensemble_pred_energy_actual))
```

```python
import random

# Initialize lists to store predictions
treated_predictions = []
energy_predictions = []
```

```python
# Number of loops needed to reach 1825 predictions
n_loops = 1825 // len(X_train_actual)  # Integer division to get full loops
remaining_rows = 1825 % len(X_train_actual)  # Remaining rows after full loops
```

```python
# Loop through and concatenate the DataFrame with itself
X_future = pd.concat([X_train_actual] * n_loops, ignore_index=True)
```

```python
# Add the remaining rows
if remaining_rows > 0:
    X_future = pd.concat([X_future, X_train_actual.iloc[:remaining_rows]], ignore_index=True)
```

```python
# Make the predictions using the GB models
pred_treated = gb_model_treated.predict(X_future)
pred_energy = gb_model_energy.predict(X_future)

# Append the predictions to the lists
treated_predictions.extend(pred_treated)
energy_predictions.extend(pred_energy)
```

```python
# Convert predictions to DataFrame
pred_df = pd.DataFrame({
    'Treated_Water_Predictions': treated_predictions,
    'Energy_Consumption_Predictions': energy_predictions
})
```

```python
# Generate a date column
last_date = pd.to_datetime(data['Date'].iloc[-1])
future_dates = pd.date_range(start=last_date, periods=len(treated_predictions) + 1, freq='D')[1:]
```
